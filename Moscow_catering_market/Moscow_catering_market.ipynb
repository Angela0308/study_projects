# Рынок заведений общественного питания Москвы

## Описание проекта
Инвесторы из фонда «Shut Up and Take My Money» решили попробовать себя в новой области и открыть заведение общественного питания в Москве. Заказчики ещё не знают, что это будет за место: кафе, ресторан, пиццерия, паб или бар, — и какими будут расположение, меню и цены. Для исследования доступен датасет с заведениями общественного питания Москвы, составленный на основе данных сервисов Яндекс Карты и Яндекс Бизнес на лето 2022 года. Информация, размещённая в сервисе Яндекс Бизнес, могла быть добавлена пользователями или найдена в общедоступных источниках, носит исключительно справочный характер.

### Задачи
* Подготовить исследование рынка Москвы
* Найти особенности, которые в будущем помогут в выборе подходящего инвесторам места
* Презентовать полученные результаты

### Описание данных
Датасет: `moscow_places.csv`. Структура: 
* `name` — название заведения; 
* `address` — адрес заведения; 
* `category` — категория заведения, например «кафе», «пиццерия» или «кофейня»; 
* `hours` — информация о днях и часах работы; 
* `lat` — широта географической точки, в которой находится заведение; 
* `lng` — долгота географической точки, в которой находится заведение; 
* `rating` — рейтинг заведения по оценкам пользователей в Яндекс Картах (высшая оценка — 5.0); 
* `price` — категория цен в заведении, например «средние», «ниже среднего», «выше среднего» и так далее; 
* `avg_bill` — строка, которая хранит среднюю стоимость заказа в виде диапазона, например: 
  * «Средний счёт: 1000–1500 ₽»; 
  * «Цена чашки капучино: 130–220 ₽»; 
  * «Цена бокала пива: 400–600 ₽». и так далее; 
* `middle_avg_bill` — число с оценкой среднего чека, которое указано только для значений из столбца avg_bill, начинающихся с подстроки «Средний счёт»: 
  * Если в строке указан ценовой диапазон из двух значений, в столбец войдёт медиана этих двух значений. 
  * Если в строке указано одно число — цена без диапазона, то в столбец войдёт это число. 
  * Если значения нет или оно не начинается с подстроки «Средний счёт», то в столбец ничего не войдёт. 
* `middle_coffee_cup` — число с оценкой одной чашки капучино, которое указано только для значений из столбца avg_bill, начинающихся с подстроки «Цена одной чашки капучино»: 
  * Если в строке указан ценовой диапазон из двух значений, в столбец войдёт медиана этих двух значений. 
  * Если в строке указано одно число — цена без диапазона, то в столбец войдёт это число. 
  * Если значения нет или оно не начинается с подстроки «Цена одной чашки капучино», то в столбец ничего не войдёт. 
* `chain` — число, выраженное 0 или 1, которое показывает, является ли заведение сетевым (для маленьких сетей могут встречаться ошибки): 0 — заведение не является сетевым 1 — заведение является сетевым 
* `district` — административный район, в котором находится заведение, например Центральный административный округ; 
* `seats` — количество посадочных мест.

## Загрузка данных

```python
# импорт библиотек
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
# загружаю данные
moscow_places = pd.read_csv('moscow_places.csv')
moscow_places.head(10)
```

```python
# проверяю пропуски и типы данных
moscow_places.info()
```

### Обзор данных
Для дальнейшей работы потребуется добавить столбцы с названием улиц и отметкой о том, работает ли заведение круглосуточно, обработать пропуски в столбцах `hours`, `seats`, `price`, `avg_bill`, `middle_avg_bill`, `middle_coffee_cup`, для удобства визуализации исследования заменить числовые значения в столбце `chain` на объектные, а также привести названия в `name` к нижнему регистру для проверки неявных дубликатов.

## Предобработка данных

```python
# проверяю наличие дубликатов
moscow_places.duplicated().sum()
```

```python
# создаю столбец street с названиями улиц
moscow_places['street'] = moscow_places['address'].apply(lambda x: x.split(', ')[1])
```

```python
# создаю столбец is_24/7 с обозначением, что заведение работает ежедневно и круглосуточно
def is_24_7(hours):
    if hours == 'ежедневно, круглосуточно':
        val = True
    else:
        val = False
    return val
moscow_places['is_24_7'] = moscow_places['hours'].apply(is_24_7)
#moscow_places.sample(5)
```

```python
# меняю числовые обозначения принадлежности к сети
moscow_places['chain'] = moscow_places['chain'].replace(1, 'сетевое').replace(0, 'не сетевое')
```

```python
# привожу значения к нижнему регистру и проверяю дубликаты
moscow_places['name'] = moscow_places['name'].str.lower()
moscow_places = moscow_places.drop([1430, 2211, 3109]) #удаляю "не сетевые" дубли 
```

### Итоги предобработки
В результате предобработки данных были обнаружены и удалены 3 неявных дубликата. Столбцы `hours`, `seats`, `price`, `avg_bill`, `middle_avg_bill`, `middle_coffee_cup`  содержат большое количество пропусков, для заполнения которых данных не хватает, поэтому оставлены без изменений. Добавлены столбцы с названием улицы `street` и информацией о том, работает ли заведение ежедневно и круглосуточно `is_24_7`. Данные в стобце `chain` заменены с числовых обозначений на текстовые.

Кафе - самый популярный тип заведений, наименее популярные - столовые и булочные.

```python
# проверяю, как распределены посадочные места
moscow_places['seats'].hist(bins=20)
plt.show()
```

```python
# проверяю размах посадочных мест
fig = px.box(moscow_places, x='seats', title='Размах количества посадочных мест')
fig.show(renderer='png')
moscow_places['seats'].describe()
```

```python
# считаю медианное значение количества посадочных мест по категориям без учета выбросов
mp_seats = moscow_places[moscow_places['seats']<=290].groupby('category')['seats'].median().sort_values(ascending=False)
display(mp_seats)
display(mp_seats.median())
```

```python
# визуализирую среднее (медианное) количество посадочных мест по категориям зеведений
fig = px.bar(x=mp_seats.index, y=mp_seats, title='Среднее количество посадочных мест',
             labels={'x':'категория', 'y': 'посадочные места'}, text=mp_seats)
fig.show(renderer='png')
```

Количество посадочных мест больше 290 встречается редко, без их учета медианное количество посадочных мест равно 65, при этом самыми вместимыми являются рестораны (80 человек) и бары/пабы (77 человек), наименьшая вместимость у булочных (48 человек).

```python
# нахожу соотношение сетевых и несетевых заведений в датасете
mp_chain = moscow_places['chain'].value_counts()
mp_chain
```

```python
# визуализирую соотношение сетевых и несетевых заведений
fig = px.pie(values=mp_chain, names=mp_chain.index, title='Соотношение сетевых и несетевых заведений',
             color = mp_chain.index, color_discrete_map={'не сетевое':'darkblue', 'сетевое':'royalblue'})
fig.show(renderer='png')
```

Несетевые заведения преобладают - почти 62% (5201 штук). Доля сетевых заведений 38% (3205 штук).

```python
# проверяю какие заведенеия чаще являются сетевыми
mp_chain_category = moscow_places.pivot_table(index=['category', 'chain'], values='name', aggfunc='count').sort_values(
    by=['chain', 'name'], ascending=False).reset_index()
display(mp_chain_category)
```

```python
# строю график соотношения сетевых и не сетевых заведений по категориям
fig = px.bar(mp_chain_category, x='category', y='name', color='chain', barmode='group',
            title='Распределение сетевых и не сетевых заведений по категориям',
            labels={'category':'категория', 'name': 'количество', 'chain': 'обозначение'}) 
fig.show(renderer='png')
```

Среди сетевых заведений самыми популярными также являются кафе, рестораны и кофейни, но в отличие от ресторанов и кафе у кофеен число сетевых заведений преобладает над не сетевыми.

```python
# формирую топ-15 названий сетей в Москве
top_15_chain_name = (
    moscow_places[moscow_places['chain'] == 'сетевое']).groupby('name')['name'].count().sort_values(ascending=False).head(15)

# визуализирую топ-15
fig = px.bar(x=top_15_chain_name.index, y=top_15_chain_name, title='Топ-15 популярных сетей в Москве',
             labels={'x':'название', 'y': 'количество'}, text=top_15_chain_name)
fig.show(renderer='png')
```

Топ-3 популярных сетей: Шоколадница, Доминос пицца, Додо пицца.

```python
# строю график соотношения категорий заведений внутри топ-15 сетей Москвы
mp_top_15_category = moscow_places[moscow_places['name'].isin(top_15_chain_name.index)].pivot_table(
    index=['name', 'category'], values='address',  aggfunc='count').sort_values(by='address', ascending=False).reset_index()
mp_top_15_category
fig = px.bar(mp_top_15_category, x='name', y='address', color='category',
             title='Категории топ-15 популярных сетей в Москве',
             labels={'name':'название', 'address': 'количество'})
fig.show(renderer='png')
```

Часть популярных сетей представлены разными типами заведений. Если внутри сети встречаются разные типы заведений, как правило, какой-то тип преобладает. Например, CofeFest: 31 кофейня и 1 кафе. 

```python
# строю график соотношения категорий заведений внутри топ-15 сетей Москвы
fig = px.bar(mp_top_15_category.groupby('category')['address'].sum().reset_index().sort_values(by='address', ascending=False),
             x='category', y='address',
             title='Соотношение категорий заведений внутри топ-15 сетей Москвы',
             labels={'category':'категория', 'address': 'количество'})
fig.show(renderer='png')
```

В популярных сетях самой распространенной категорией является кофейня, следующие по популярности рестораны почти в 2 раза меньше. Самые распространненные в общей массе кафе только на 4м месте.

```python
# визуализирую общее количество заведений по административным районам Москвы
fig = px.bar(x=moscow_places['district'].value_counts().index, y=moscow_places['district'].value_counts(),
             title='Количество заведений по районам',
             labels={'x':'район', 'y': 'количество'}, text=moscow_places['district'].value_counts())
fig.show(renderer='png')
```

```python
# визуализирую как распределяются категории заведений по районам
fig = px.bar(moscow_places.groupby(['district', 'category'])['name'].count().reset_index(),
             x='name', y='district', color='category',
             title='Количество заведений каждой категории по районам',
             labels={'name':'количество', 'district': 'район', 'category': 'обозначения'})
fig.show(renderer='png')
```

Количество заведений общественного питания в ЦАО ожидаемо больше, чем в других районах. Северо-Запад значительно уступает, по остальным районам распределение более равномерное. Почти половина баров и пабов города сосредоточены в центре. Количество кафе превышает количество ресторанов везде, кроме центра.

```python
# визуализирую распределение средних рейтингов по категориям заведений
fig = px.bar(moscow_places.groupby('category', as_index=False)['rating'].mean().round(1).sort_values(by='rating', ascending=False),
             x='category', y='rating',
             title='Распределение средних рейтингов по категориям заведений',
             labels={'category':'категория', 'rating': 'оценка'}, text='rating')
fig.show(renderer='png')
```

Различия в средней оценке между категориями не значительные, находятся в диапазоне от 4.1 до 4.4.

```python
# вычисляю средний рейтинг заведений по районам
mp_district_rating = moscow_places.groupby('district', as_index=False)['rating'].mean().round(1)
mp_district_rating
```

Средний рейтинг заведений по районам отличается слабо. Самый высокий у ЦАО - 4.4, самый низкий у Северо-Востока и Юго-Востока - 4.1, у всех остальных округов средний рейтинг 4.2.

```python
# определяю топ-15 улиц по количеству заведений
top_street = moscow_places.pivot_table(index='street', values='category', aggfunc='count').sort_values(by='category', ascending=False).head(15).reset_index()
top_15_street = moscow_places[moscow_places['street'].isin(top_street['street'])]
```

```python
# визуализирую распределение количества заведений и их категорий по топовым улицам 
fig = px.bar(top_15_street.groupby(['street', 'category'])['name'].count().reset_index(),
             x='name', y='street', color='category',
             title='Количество заведений каждой категории по улицам (топ-15)',
             labels={'name':'количество', 'street': 'улица'})
fig.show(renderer='png')
```

Больше всего заведений питания сосредоточено на проспекте Мира, там же самое большое количество кафе, кофеен и ресторанов.

```python
# визуализирую распределение категорий на улицах с одним заведением
fig = px.bar(x=one_str_cat['category'].value_counts().index,
             y=one_str_cat['category'].value_counts(), title='Распределение категорий на улицах с одним заведением',
             labels={'x':'категория', 'y': 'количество'}, text=one_str_cat['category'].value_counts())
fig.show(renderer='png')
```

Распределение одиночных заведений на улице совпадает с общим трендом: чаще всего это кафе, на втором месте рестораны, на третьем кофейни.

```python
# вычисляю средний чек по районам
mp_district_avg_bill = moscow_places.groupby('district', as_index=False)['middle_avg_bill'].median()
mp_district_avg_bill
```

Средний чек ЦАО ожидаемо оказался самым высоким, на уровне с ним Западный административный округ, куда входят аэропорт Внуково, "Сколково" и Рублевка. Обед на Юго-Востоке обойдется дешевле всего.
